```python
import matplotlib.pyplot as plt
import numpy as np

# Data points
rate = [5, 10, 20, 33.3, 100]
mqtt = [5.01, 9.97, 19.9, 32.15, 91.99]
serial = [4.9612, 9.7458, 18.88, 30.43, 64.31]

# Create the figure and axis
plt.figure(figsize=(10, 6))

# Plot MQTT data with a curve
plt.plot(rate, mqtt, 'bo-', label='MQTT', marker='o')

# Plot Serial data with a curve
plt.plot(rate, serial, 'rs-', label='Serial', marker='s')

# Add labels and title
plt.xlabel('Rate of data transfer from arduino (1 / delay)')
plt.ylabel('Rate of data receipt from laptop')
plt.title('Comparison of MQTT and Serial Data Rates')
plt.grid(True)
plt.legend()

# Adjust the y-axis to start from 0
plt.ylim(bottom=0)

# Display the plot
plt.tight_layout()
plt.show()
```
